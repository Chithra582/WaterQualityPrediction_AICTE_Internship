```python
# import the necessary libraries
import pandas as pd
import numpy as np

from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# load the dataset
df=pd.read_csv('PB_All_2000_2021.csv', sep=';')
df
```

```python
df.info()  # dataset info
```

```python
# rows and cols
df.shape
```

```python
# statistics of the data
df.describe().T
```

```python
# missing values
df.isnull().sum()
```

```python
# date is in object - date format
df['date']=pd.to_datetime(df['date'], format='%d.%m.%Y')
df
```

```python
df.info()
```

```python
df=df.sort_values(by=['id', 'date'])
df.head()
```

```python
df['year']=df['date'].dt.year
df['month']=df['date'].dt.month
```

```python
df.head()
```

```python
df.columns
```

```python
pollutants=['O2','NO3','NO2','SO4','PO4','CL']
df[pollutants].isnull().sum() 
```

```python
pollutants=['O2','NO3','NO2','SO4','PO4','CL']
```

```python
df=df.dropna(subset=pollutants)
df.head()
```

```python
df.isnull().sum()
```

```python
X=df[['id','year']]
y=df[pollutants]
```

```python
#encoding - onehotencoder -22 stations - 1 - 1
X_encoded=pd.get_dummies(X, columns=['id'], drop_first=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(
    X_encoded,y,test_size=0.2, random_state=42
) #Train, test and split
```

```python
model=MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=32))
model.fit(X_train, y_train) #train the model
```

```python
y_pred = model.predict(X_test) #evaluate the model
```

```python
print("Model Performance on the Test Data:")
for i, pollutant in enumerate(pollutants):
    print(f'{pollutant}:')
    print('   MSE:', mean_squared_error(y_test.iloc[:, i],y_pred[:, i]))
    print('   R2:', r2_score(y_test.iloc[:, i],y_pred[:, i]))
    print()
```

```python
station_id='22'
year_input=2024

input_data=pd.DataFrame({'year':[year_input], 'id':[station_id]})
input_encoded=pd.get_dummies(input_data, columns=['id'])

missing_cols=set(X_encoded.columns)- set(input_encoded.columns)
for col in missing_cols:
    input_encoded[col]=0
input_encoded=input_encoded[X_encoded.columns]

predicted_pollutants=model.predict(input_encoded)[0]

print(f"\nPredicted pollutant levels for station '{station_id}' in {year_input}:")
for p,val in zip(pollutants, predicted_pollutants):
    print(f"  {p}:  {val:.2f}")
```

```python
import matplotlib.pyplot as plt

#Visualizing the predicted pollutant levels
plt.figure(figsize=(10,5))
plt.bar(pollutants, predicted_pollutants, color='skyblue')
plt.xlabel('pollutants')
plt.ylabel('predicted Levels')
plt.title(f"predicted pollutant levels for station {station_id} in {year_input}")
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()
```
